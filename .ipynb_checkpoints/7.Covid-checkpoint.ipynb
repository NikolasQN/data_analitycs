```python
# импортирование необходимых библиотек
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# 1-мерный массив (вектор-строка)
x_arr = np.arange(10)
x_arr
```

```python
y_arr = x_arr

# визуализация
plt.plot(x_arr, y_arr)

plt.show()
```

```python
# 1-мерный массив из 100 значений
# из линейной последовательности в диапазоне от 0 до 10

x = np.linspace(-10,10,100)
x
```

```python
# функция гиперболический тангенс
y0 = np.tanh(x)

# функция синусоиды
y1 = np.sin(x)

# функция сигмоиды
def sigmoid(arr):
    return 1 / (1+np.exp(-arr))

y2 = sigmoid(x)

# визуализация
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```

```python
# загрузка данных
data_0 = pd.read_csv("data/covid_19_data.csv")
data_0.head()
```

```python
data_0.info()
```

```python
data_0.iloc[-1, :]
```

```python
data_0.columns
```

```python
# удалаяем ненужные столбцы
data_1 = data_0.drop(['SNo', 'Last Update'], axis=1)
data_1.head()
```

```python
# проверка на дубликаты
data_1.duplicated(['ObservationDate', 'Province/State', 'Country/Region', 
                   'Confirmed', 'Deaths', 'Recovered']).sum()
```

```python
# удаление дубликатов
data_1.drop_duplicates(['ObservationDate', 'Province/State', 'Country/Region', 
                   'Confirmed', 'Deaths', 'Recovered'], inplace=True)
```

- анализ

```python
# изучение списка стран
data_1['Country/Region'].unique()
```

```python
data_1.columns
```

```python
# изучаем страну "Others"
ship_data = data_1.loc[data_1['Country/Region'] == "Others", :]
ship_data
```

```python
data_1.columns
```

```python
# визуализация по кораблю
plt.figure(figsize=(12, 6))

plt.plot(ship_data["ObservationDate"], ship_data['Confirmed'])
plt.plot(ship_data["ObservationDate"], ship_data['Deaths'])
plt.plot(ship_data["ObservationDate"], ship_data['Recovered'])

plt.legend(["Заболевшие", "Погибшие", "Выздоровившие"])
plt.xlabel("даты")
plt.ylabel("кол-во людей")
plt.title("Динамика Covid-19 на кораблей 'Diamond Princess'")
plt.show()
```

```python
# конвертация дат
data_1["ObservationDate"][0]
```

```python
#  метод для конвертации временных данных
data = '01/22/2020'
data = '11th of September 2022'
pd.to_datetime(data)
```

```python
data_2= data_1.copy()

data_2["ObservationDate"] = pd.to_datetime(data_2["ObservationDate"])

data_2["ObservationDate"][0]
```

```python
# поэлементное извлечение даты (без часов минут и секунд)
data_2["ObservationDate"] = data_2["ObservationDate"].apply(lambda d: d.date())

data_2["ObservationDate"][0]
```

```python
data_2.columns
```

```python
data_2['Country/Region'].str
```

```python
data_2['Country/Region'].unique()
```

```python
# убираем артефакты из названий стран
data_2['Country/Region'] = data_2['Country/Region'].str.strip(" (),''")
data_2['Country/Region'].unique()
```

```python
data_2.columns
```

```python
# группировка данных по странам и датам
cd_data = data_2.groupby(['Country/Region', 'ObservationDate']).sum()
cd_data.head(20)
```

```python
# данные макисмумов по странам 
max_data = cd_data.groupby(['Country/Region'])[['Confirmed',
       'Deaths', 'Recovered']].max()
max_data.head()
```

```python
max_data.sort_values(by='Confirmed', ascending=False).head(10)
```

```python
plt.plot(cd_data["Country/Region"], ship_data['Confirmed'])
plt.plot(cd_data["Country/Region"], ship_data['Deaths'])
plt.plot(cd_data["Country/Region"], ship_data['Recovered'])

plt.legend(["Заболевшие", "Погибшие", "Выздоровившие"])
plt.xlabel("даты")
plt.ylabel("кол-во людей")
plt.title("Динамика Covid-19 в США'")
plt.show()
```
